#Parte II do Trabalho

```python
#Configurações iniciais para o ambiente
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
import numpy as np
sns.set()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/JeffersonEstevo/Projeto-Ciencia-de-Dados/main/avaliacaodocencia.csv') #Data frame inicial
df.info() #Exibir dataframe inicial
df1 = df.dropna() #Retirando valores nulos
df1.info()
df1 #Exibir dataframe ajustado
```

```python
#Criando Data frames com avaliações nos anos 2020, 2021, 2022

# avaliações de 2020
df_2020 = df1.loc[df1['ano'] == 2020, 
 ['nome_docente', 'ano',  'qtd_discentes', 'postura_profissional_media', 
  'atuacao_profissional_media', 'autoavaliacao_aluno_media']]

# avaliações de 2021
df_2021 = df1.loc[df1['ano'] == 2021, 
 ['nome_docente', 'ano',  'qtd_discentes', 'postura_profissional_media', 
  'atuacao_profissional_media', 'autoavaliacao_aluno_media']]

# avaliações de 2022
df_2022 = df1.loc[df1['ano'] == 2022, 
 ['nome_docente', 'ano',  'qtd_discentes', 'postura_profissional_media', 
  'atuacao_profissional_media', 'autoavaliacao_aluno_media']]

df_2020.info()
df_2021.info()
df_2022.info()
```

```python
#Docentes 2020 ordenados por nome do professor
df_2020_sort = df_2020.sort_values(by=['nome_docente'], ascending=True)
df_2020_sort
```

```python
#Docentes 2021 ordenados por nome do professor
df_2021_sort = df_2021.sort_values(by=['nome_docente'], ascending=True)
df_2021_sort
```

```python
#Docentes 2022 ordenados por nome do professor
df_2022_sort = df_2022.sort_values(by=['nome_docente'], ascending=True)
df_2022_sort
```

```python
#10 melhores professores de 2020

#Ordenação por postura profissional
a1 = df_2020.sort_values(by=['postura_profissional_media'], ascending=False)
df_2020_sort_n = a1.head(10)['nome_docente']
df_2020_sort_n
```

```python
#10 melhores professores de 2021

#Ordenação por postura profissional
a2 = df_2021.sort_values(by=['postura_profissional_media'], ascending=False)
df_2021_sort_n = a2.head(10)['nome_docente']
df_2021_sort_n
```

```python
#10 melhores professores de 2022

#Ordenação por postura profissional
a3 = df_2022.sort_values(by=['postura_profissional_media'], ascending=False)
df_2022_sort_n = a3.head(10)['nome_docente']
df_2022_sort_n
```

#Geração de gráficos

```python
# Informações para geração dos gráficos

#Médias das avaliações

#Obtendo postura média dos professores no ano de 2020
media_postura_2020 = df_2020['postura_profissional_media'].mean()
#Obtendo postura média dos professores no ano de 2021
media_postura_2021 = df_2021['postura_profissional_media'].mean()
#Obtendo postura média dos professores no ano de 2022
media_postura_2022 = df_2022['postura_profissional_media'].mean()

#Obtendo atuação média dos professores no ano de 2020
media_atuacao_2020 = df_2020['atuacao_profissional_media'].mean()
#Obtendo atuação média dos professores no ano de 2021
media_atuacao_2021 = df_2021['atuacao_profissional_media'].mean()
#Obtendo atuação média dos professores no ano de 2022
media_atuacao_2022 = df_2022['atuacao_profissional_media'].mean()

#Obtendo autoavaliação média dos alunos no ano de 2020
media_autoavaliacao_2020 = df_2020['autoavaliacao_aluno_media'].mean()
#Obtendo autoavaliação média dos alunos no ano de 2021
media_autoavaliacao_2021 = df_2021['autoavaliacao_aluno_media'].mean()
#Obtendo autoavaliação média dos alunos no ano de 2022
media_autoavaliacao_2022 = df_2022['autoavaliacao_aluno_media'].mean()
```

```python
import plotly.express as px #Plotly express
import plotly.graph_objs as go #Go
```

```python
anos = [2020, 2021, 2022]
posturas = [media_postura_2020, media_postura_2021, media_postura_2022]

fig = px.bar(x=anos, y=posturas, barmode="group", labels={'x': 'Anos', 'y': 'Posturas'}, color=posturas)
fig.update_layout(title='Postura por ano')
fig.show()
```

```python
anos = [2020, 2021, 2022]
atuacoes = [media_atuacao_2020, media_atuacao_2021, media_atuacao_2022]

fig = px.bar(x=anos, y=atuacoes, barmode="group", labels={'x': 'Anos', 'y': 'Atuações'}, color=atuacoes)
fig.update_layout(title='Atuação por ano')
fig.show()
```

```python
anos = [2020, 2021, 2022]
autoavaliacoes = [media_autoavaliacao_2020, media_autoavaliacao_2021, media_autoavaliacao_2022]

fig = px.bar(x=anos, y=autoavaliacoes, barmode="group", labels={'x': 'Anos', 'y': 'Autoavaliações'}, color=autoavaliacoes)
fig.update_layout(title='Autoavaliaçõeso por ano')
fig.show()
```
